# Kobe Shot Selection
In this dataset we have a record of every shot that Kobe Bryant took in his long career with the LA Lakers. Details about the nature of each shot are included such as location on the court, shot type, and shot zone area. Time information is also included to contextualize the shot such as the time left in the game, and date. 

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.columns.values
```

```python
df['home_game'] = df['matchup'].apply(lambda x: 1 if '@' in x else 0)
```

A big thing in basically is momentum and being 'in the zone'. This is especially evident with star players as when they start to score and shoot well, their team will push the ball towards them to let them take even more shots.

```python
# def streakShotPerc(df, n_prev_shots=5):
    
game_grouped = df.groupby('game_id')
df['streak'] = game_grouped['shot_made_flag'].apply(lambda df: df.rolling(5).mean())
```

```python
court_ratio = 520/500
fig, axs = plt.subplots(1, 2, figsize=(8*2, 8*court_ratio))
df[(df['loc_y'] <= 470)].plot.hexbin(x='loc_x', y='loc_y',
                                     gridsize=52, ax=axs[0], grid=False, 
                                     alpha=0.7, vmax=50, cmap=plt.cm.jet, colorbar=True,
                                     title='Total Shots Taken', xticks=[], yticks=[], use_index=False)
df[(df['loc_y'] <= 470)].plot.hexbin(x='loc_x', y='loc_y', C='shot_made_flag',
                                     reduce_C_function=lambda x: sum(x)/len(x) * 100,
                                     gridsize=52, ax=axs[1], grid=False, 
                                     alpha=0.7, cmap=plt.cm.jet, colorbar=True,
                                     title='Shot Percentage', xticks=[], yticks=[])
for ax in axs:
    ax.set_xlim([-250, 250]); ax.set_ylim([-50, 470])
    ax.imshow(mpimg.imread('court.png'), extent=[-250, 252, -50, 890])
    ax.set_xlabel(''); ax.set_ylabel('')
cb1 = fig.get_axes()[2]
cb1_cm_labels = cb1.get_yticklabels()
cb1_cm_labels[-1].set_text(cb1_cm_labels[-1].get_text() + '+')
cb1.set_yticklabels(cb1_cm_labels)
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
# colors = ['b', 'g', 'r']
colors = plt.cm.rainbow(np.linspace(0,1,5))
for shot_type, c in zip(df['combined_shot_type'].unique(), colors):
    df[df['combined_shot_type']==shot_type].plot.scatter(x='loc_x', y='loc_y', color=c, label=shot_type, ax=ax);

# df.plot.scatter(x='c', y='d', color='DarkGreen', label='Group 2', ax=ax);
```
